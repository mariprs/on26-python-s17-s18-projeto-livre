<h1 style='font-family: Lucida Console; text-align:center; font-weight: 800'>Descrição:</h1>
<ol><p style="text-align: justify">Este projeto tem como tema a desigualdade de gênero e visa construir visualizações do Brasil através da análise de dois datasets:
<li><ul style="text-align: justify;"><b>Global Gender Gap Report:</b> ranking de desigualdade de gênero publicado anualmente desde 2005 pelo Fórum Econômico Mundial. Uma explicação melhor de como os valores do dataset são calculados se encontra abaixo.
O índice examina quatro áreas de desigualdade entre homens e mulheres em 130 economias ao redor do mundo, compondo mais de 93% da população mundial: 
<div style="margin-left: 50px;">
<li><b>Oportunidade e participação econômica</b>: salários, níveis de participação e acesso a empregos considerados de alta habilidade.</li> 
<li><b>Obtenção educacional</b>: acesso ao ensino básico e ao superior.</li>
<li><b>Empoderamento político</b>: representação em estruturas políticas.</li>
<li><b>Saúde e sobrevivência</b>: expectativa de vida e razão de natalidade (nesse caso, como há menos nascimentos femininos que masculinos, sendo 944 para 1000, leva-se em conta que mulheres vivem ao menos 6% a mais que homens - caso o número seja menor que 6%, é assumido como parte da disparidade).</li></p></ul></div>
<p style="text-align: justify"><li><b><ul style="text-align: justify;">Gender Inequality Index:</b> índice para medir desigualdade de gênero introduzido em 2010 pelo Programa das Nações Unidas para o Desenvolvimento. Surgiu com o objetivo de remediar falhas de índices similares da ONU, o Gender Development Index (GDI) e o Gender Empowerment Measure (GEM), ambos introduzidos em 1995.
O índice examina três eixos: 
<div style="margin-left: 50px;">
<li><b>Saúde reprodutiva</b>: através da observação do índice de mortalidade materna e do índice de natalidade adolescente.</li> 
<li><b>Empoderamento</b>: porcentagem de mulheres presentes no plenado e porcentagem de mulheres e homens que atingem o ensino superior</li>
<li><b>Participação no mercado de trabalho</b>: medida pela participação feminina no mercado de trabalho (e a comparação com a masculina).</li></ol>
</p></ul></div>



<h2 style='font-family: Lucida Console; text-align:center; font-weight: 800'>Saiba mais sobre o Gender Inequality Index:</h2>
<ol><p style="text-align: justify">Este banco de dados é consiste das seguintes colunas:</p>
<li><b>Country</b>: o país ao qual os dados se referem.</li>
<li><b>Human Development</b>: o nível de desenvolvimento humano do país.</li>
<li><b>GII</b>: a pontuação do Índice de Desigualdade de Gênero.</li>
<li><b>Rank</b>: a posição do país no ranking global de desigualdade de gênero. <b>Este é o único valor que irei usar no projeto.</b></li>
<li><b>Maternal Mortality</b>: a taxa de mortalidade materna, que funciona como indicador crítico de saúde feminina. Foi calculado um ratio de mortes por 100,000 nascimentos.</li>
<li><b>Adolescent Birth Rate</b>: a taxa de natalidade na adolescência, fator que afeta saúde, oportunidades econômicas e reflete na educação.</li>
<li><b>Seats in Parliament</b>: a representação feminina no parlamento em porcentagem, um fator de igualdade política.</li>
<li><b>Female Secondary Education</b>: a participação feminina no ensino médio. Em porcentagem e incluído apenas mulheres de 25 anos e mais velhas.</li>
<li><b>Male Secondary Education</b>: a participação masculina no ensino médio. Em porcentagem e incluído apenas homens de 25 anos e mais velhos.</li>
<li><b>Female Labor Force Participation</b>: a participação feminina no mercado de trabalho. Em porcentagem, incluso idades de 15 para cima.</li>
<li><b>Male Labor Force Participation</b>: a participação masculina no mercado de trabalho.Em porcentagem, incluso idades de 15 para cima.</li></ol>

<h4 style='font-family: Lucida Console; text-align:center; font-weight: 800'>Processo do Cálculo:</h4>
<ol><p style="text-align: justify">O relatório segue uma metodologia específica para calcular a disparidade de gênero, incluindo:</p>
<li><b>Tratamento de Zeros e Valores Extremos</b>: os dados são ajustados, com valores de mortalidade materna e natalidade adolescente sistematicamente limitados entre 10 e 1.000. Países com representação parlamentar relativa a 0 são considerados como 0,1, levando em conta que mulheres têm alguma influência política e que a média geométrica não pode ter valor 0.</li>
<li><b>Agregação por Médias Geométricas</b>: os indicadores são agregados usando médias geométricas, tornando o relatório associativo e sensível à distribuição das lacunas de gênero.</li>
<li><b>Agregação entre Grupos de Gênero por Média Harmônica</b>: os índices femininos e masculinos são combinados usando a média harmônica das médias geométricas para capturar a desigualdade entre gêneros.</li>
<li><b>Cálculo do Índice de Desigualdade de Gênero</b>: o índice final é obtido comparando o índice de gênero igualmente distribuído com um padrão de referência, utilizando a metodologia descrita acima.</li></ol>

<h3 style='font-family: Lucida Console; font-weight: 800;text-align: center;'>E quais são os valores do Brasil do próprio Report de Gap de Gênero?</h3>

```python
#Aqui eu resolvi isolar os indicadores que o Brasil não tem pra focar apenas nos dados que eu posso exibir
indicadores = ["Proporção de tempo gasto com trabalho doméstico e de cuidado não remunerado, mulheres (% do dia de 24 horas)",
               "Proporção de tempo gasto com trabalho doméstico e de cuidado não remunerado, homens (% do dia de 24 horas)"]

# Verificar se ambos os indicadores têm valor zero. Usei a função iloc. Aqui, os dois pontos : indicam que estamos selecionando todas as linhas do dataframe e a partir da quarta coluna até a última coluna (3: após a vírgula). Isso significa que estamos pegando todas as linhas, mas só as colunas a partir da quarta posição até o final.
condicao_zeros = (df_gender_gap['Nome do Indicador'].isin(indicadores)) & (df_gender_gap.iloc[:, 3:] == 0).all(axis=1)

# Remover a linha em que ambos os indicadores têm valor zero para todos os anos do Brasil
df_gender_gap = df_gender_gap[~condicao_zeros]
```

<h3 style='font-family: Lucida Console; font-weight: 800;text-align: center;'>Como se deu o desemprego na época de Economia no Brasil? E em outras partes da América Latina? E no Caribe?</h3>

```python
# Isolando apenas os indicadores que eu quero
desemprego_indices = [
    'Desemprego, homens (% da força de trabalho masculina) (estimativa modelada pela OIT)',
    'Desemprego, mulheres (% da força de trabalho feminina) (estimativa nacional)'
]

# Filtrando o dataframe
desemprego = df_gender_gap[df_gender_gap['Nome do Indicador'].isin(desemprego_indices)]

# Resetar o índice para facilitar a pivotagem
desemprego = desemprego.reset_index(drop=True)

# Transpor os dados para ter os anos como colunas
desemprego_pivot = desemprego.T.reset_index()
 
# Usar a primeira linha como cabeçalho de coluna
desemprego_pivot.columns = desemprego_pivot.iloc[0]

# Remover a primeira linha (pois agora ela é o cabeçalho)
desemprego_pivot = desemprego_pivot[1:]

# Renomear as colunas para nomes mais apropriados
desemprego_pivot = desemprego_pivot.rename(columns={'Nome do País': 'Ano', 'Código do País': 'Nome do País'})

# Removendo as duas primeiras linhas do dataframe
desemprego_pivot = desemprego_pivot[2:]


# Filtrar os dados para Brasil
brasil = desemprego_pivot[['Ano', 'Brasil']]

# Usar melt para transformar os anos em uma coluna 'Ano' e 'Valor' com os valores de desemprego
desemprego_melted = brasil.melt(id_vars='Ano', var_name='Indicador', value_name='Valor')

# Selecionar apenas os anos de interesse (por exemplo, 2019, 2020, 2021)
anos_interesse = ['2019', '2020', '2021']
desemprego_interesse = desemprego_melted[desemprego_melted['Ano'].isin(anos_interesse)]

desemprego_pandemia = desemprego_interesse.groupby(['Ano', 'Indicador']).sum().reset_index()

# Somando ambos
desemprego_melted['Indicador'] = desemprego_melted['Indicador'].str.replace(r"\s*\(%.*\)", "", regex=True)
soma_desemprego = desemprego_melted.groupby(['Ano', 'Indicador']).sum().reset_index()

# Plotar o gráfico de linha para o desemprego no Brasil nos anos de interesse (somando homens e mulheres)
plt.figure(figsize=(10, 6))
sns.lineplot(data=desemprego_pandemia, x='Ano', y='Valor', hue='Indicador', marker='o')
plt.title('Demanda para trabalho durante a Pandemia (2019, 2020, 2021)')
plt.xlabel('Ano')
plt.ylabel('Desemprego (%)')
plt.legend(title='Indicador')
plt.grid(True)
plt.show()
```
